## CORPUS (2)

The sequence of word tokens in the corpus, indexed by their location in the corpus and document structures.

- UVA Box URL:https://virginia.box.com/s/30mhsvycjc8gtrn9el93mwluo24skovb
- GitHub URL for notebook used to create:https://github.com/Poptartmeow/DS5001-I-Ching-the-Book-of-Change/blob/main/finalproject_shawn.ipynb
- Delimitter: |
- Number of observations Between (should be >= 500,000 and <= 2,000,000 observations.): 53,000 tokens
- OHCO Structure (as delimitted column names): doc_id, sent_num, token_num
- Columns (as delimitted column names, including `token_str`, `term_str`, `pos`, and `pos_group`):token_str, term_str, pos, pos_group

```python
# --- CREATE CORPUS TABLE ---
def tokenize_corpus(df):
    token_data = []
    for doc_id, row in df.iterrows():
        sentences = sent_tokenize(row['content'])
        for s_idx, sent in enumerate(sentences):
            tokens = word_tokenize(sent)
            for t_idx, token in enumerate(tokens):
                token_data.append([doc_id, s_idx, t_idx, token])
    
    corpus = pd.DataFrame(token_data, columns=['doc_id', 'sent_num', 'token_num', 'token_str'])
    corpus.set_index(['doc_id', 'sent_num', 'token_num'], inplace=True)
    
    # Normalization
    corpus['term_str'] = corpus['token_str'].str.lower().replace(r'[\W_]+', '', regex=True)
    corpus = corpus[corpus['term_str'] != '']
    
    # POS Tagging
    pos_list = nltk.pos_tag(corpus['token_str'].tolist())
    corpus['pos'] = [p[1] for p in pos_list]
    corpus['pos_group'] = corpus['pos'].str[:2]
    
    return corpus

corpus = tokenize_corpus(df)
corpus.to_csv('CORPUS.csv', sep='|')

corpus.head()
```

```python
vocab = corpus['term_str'].value_counts().to_frame('n')
vocab.index.name = 'term_str'
vocab['p'] = vocab['n'] / vocab['n'].sum()
vocab['i'] = -np.log2(vocab['p']) # Information (Entropy)

# 2. Add Stopwords and Stems
sw = stopwords.words('english')
ps = PorterStemmer()
vocab['stop'] = vocab.index.isin(sw).astype(int)
vocab['porter_stem'] = [ps.stem(word) for word in vocab.index]

# 3. Add POS metadata
# We find the most frequent POS tag for each word in the CORPUS
pos_counts = corpus.groupby(['term_str', 'pos']).size().unstack(fill_value=0)
vocab['max_pos'] = pos_counts.idxmax(axis=1)
vocab['max_pos_group'] = vocab['max_pos'].str[:2] # Usually NN, VB, JJ, etc.

# 4. Add TFIDF (DFIDF)
# DFIDF is the sum of TFIDF values across all documents
# We calculate this using the original content to maintain document context
tfidf_engine = TfidfVectorizer(use_idf=True, stop_words='english')
tfidf_matrix = tfidf_engine.fit_transform(df['content'])
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(), 
    index=df.index, 
    columns=tfidf_engine.get_feature_names_out()
)
# Map the sum of TFIDF to the vocab table
vocab['dfidf'] = tfidf_df.sum()
vocab['dfidf'] = vocab['dfidf'].fillna(0) # For words filtered out by stop_words

# 5. Add N-gram length
vocab['ngram_length'] = 1

# 6. Save VOCAB
vocab.to_csv('VOCAB.csv', sep='|')

# Display top 20 by DFIDF for your markdown section
print("Top 20 Significant Words by DFIDF:")
print(vocab.sort_values('dfidf', ascending=False).head(20).index.tolist())
```

## TFIDF (3)

A Document-Term matrix with TFIDF values.

- UVA Box URL:https://virginia.box.com/s/30mhsvycjc8gtrn9el93mwluo24skovb
- UVA Box URL of DTM or BOW used to create:https://virginia.box.com/s/nn87x93fmz3yez7dywmnw5s4jp6vt1bq
- GitHub URL for notebook used to create:https://github.com/Poptartmeow/DS5001-I-Ching-the-Book-of-Change/blob/main/finalproject_shawn.ipynb
- Delimitter: |
- Description of TFIDIF formula ($\LaTeX$ OK): TF = raw count / document length; IDF = log(N / df); TFIDF = TF * IDF; then row-wise L2 normalization.

```python
# Compute TF
TF = DTM.div(DTM.sum(axis=1), axis=0)
# IDF
N = len(DTM)
DF = (DTM > 0).sum(axis=0)
IDF = np.log(N / DF)
# TF-IDF
TFIDF = TF * IDF
# L2 normalization
TFIDF_L2 = normalize(TFIDF, norm='l2')
TFIDF_L2 = pd.DataFrame(TFIDF_L2, index=DTM.index, columns=DTM.columns)
TFIDF_L2.to_csv('TFIDF.csv', sep='|')
TFIDF_L2.head()
```

## Reduced and Normalized TFIDF_L2 (3)

A Document-Term matrix with L2 normalized TFIDF values.

- UVA Box URL:https://virginia.box.com/s/30mhsvycjc8gtrn9el93mwluo24skovb
- UVA Box URL of source TFIDF table:https://virginia.box.com/s/85pacwellkhlggwkn0ayg9cceihrev7t
- GitHub URL for notebook used to create:https://github.com/Poptartmeow/DS5001-I-Ching-the-Book-of-Change/blob/main/finalproject_shawn.ipynb
- Delimitter: |
- Number of features (i.e. significant words): 1908
- Principle of significant word selection: Retain only terms that appear in at least 2 documents (document frequency ≥ 2) and are not in the sklearn English stopwords list. This removes extremely rare words (which add noise) and common function words (which carry little semantic meaning), ensuring that the subsequent PCA and LDA models focus on meaningful, content-bearing vocabulary.

```python
keep_terms = [col for col in DTM.columns if col not in ENGLISH_STOP_WORDS and (DTM[col] > 0).sum() >= 2]
DTM_filtered = DTM[keep_terms]
TF_filtered = DTM_filtered.div(DTM_filtered.sum(axis=1), axis=0)
DF_filtered = (DTM_filtered > 0).sum(axis=0)
IDF_filtered = np.log(N / DF_filtered)
TFIDF_filtered = TF_filtered * IDF_filtered
TFIDF_L2 = normalize(TFIDF_filtered, norm='l2')
TFIDF_L2 = pd.DataFrame(TFIDF_L2, index=DTM_filtered.index, columns=DTM_filtered.columns)
TFIDF_L2.to_csv('TFIDF_L2.csv', sep='|')
print(f"TFIDF_L2 shape: {TFIDF_L2.shape} (features: {TFIDF_L2.shape[1]})")
```

```python
stop_set = set(ENGLISH_STOP_WORDS)
DTM_LDA = DTM.loc[:, ~DTM.columns.isin(stop_set)]
DTM_LDA = DTM_LDA.loc[:, DTM_LDA.sum() > 5]   # min document frequency = 6

# Keep only nouns and verbs (based on VOCAB table)
v_filter = vocab[vocab['max_pos_group'].isin(['NN', 'VB'])].index
DTM_Filtered = DTM_LDA[DTM_LDA.columns.intersection(v_filter)]

# Fit LDA
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
THETA = pd.DataFrame(lda.fit_transform(DTM_Filtered), index=DTM_Filtered.index)
THETA.columns = [f"T{str(i).zfill(2)}" for i in range(n_topics)]

# PHI (topic-term matrix)
PHI = pd.DataFrame(lda.components_, columns=DTM_Filtered.columns)
PHI.index = [f"T{str(i).zfill(2)}" for i in range(n_topics)]

# Save
THETA.to_csv('THETA.csv', sep='|')
PHI.to_csv('PHI.csv', sep='|')

# Print top 5 words per topic
for i in range(n_topics):
    t_id = f"T{str(i).zfill(2)}"
    top_words = PHI.loc[t_id].sort_values(ascending=False).head(5).index.tolist()
    print(f"{t_id}: {top_words}")
```
